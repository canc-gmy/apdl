# Fourier Analysis from CSV
Spectral analysis of strain fields generated from simulation data.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
from analysis.Analysis import Analysis

# Add project root to path to import src modules
sys.path.append(os.path.abspath('..'))

try:
    plt.rcParams.update({
        "text.usetex": True,
        "font.family": "serif",
        "text.latex.preamble": r"\usepackage{amsmath}",
    })
except Exception as e:
    print(f"LaTeX fallback: {e}")
```

```python
CSV_FILE_PATH = r"../sim/sim_0.006_0.024.parquet"
ana = Analysis(CSV_FILE_PATH)

def plot_spectrum(ft_data, kx, ky, title):
    plt.figure(figsize=(10, 8))
    plt.pcolormesh(kx, ky, np.log10(1 + np.abs(ft_data)), cmap='plasma', shading='auto')
    plt.colorbar(label='$\\log_{10}(1 + |\\mathrm{FFT}|)$')
    plt.xlabel('Wavenumber $k_x$')
    plt.ylabel('Wavenumber $k_y$')
    plt.title(title)
    plt.show()

plot_spectrum(ana.ft_eyy, ana.kx, ana.ky, 'Spectrum of Deflection $w$')
```
